```
# Find Pi to the Nth Digit
# Based on Chudnovsky's algorithm https://en.wikipedia.org/wiki/Chudnovsky_algorithm
def factorial(n):
  if n < 1:
    return 1
  else:
    return n * factorial(n-1)

def findPi2(digits):
  getcontext().prec = 10000
  test = Decimal(0)
  for i in range(digits):
    Mq = Decimal(factorial(6*i)) / Decimal(factorial(3*i)) / Decimal(factorial(i)**3)
    Lq = 545140134 * i + 13591409
    Xq = (-262537412640768000)**i
    test += Decimal(Mq * Lq / Xq)
  return test

def findPi(digits): 
    C = 426880 * Decimal(10005).sqrt()
    # only need 1 iteration to get 
    f8 = findPi2(digits)
    pi = C / f8
    if digits > 1 :
      return str(pi)[0] + "." + str(pi)[2:digits+1]
    else:
      return str(pi)[0] 
```

```
# Factorial function
def factorial(n):
  if n < 1:
    return 1
  else:
    return n * factorial(n-1)

# Find e to the Nth Digit
def finde(length):
  getcontext().prec = 100
  test = Decimal(0)
  for i in range(20):
    test += Decimal(2*i + 2) / factorial(2*i + 1)

  if length > 1 :
    return str(test)[0] + "." + str(test)[2:length+1]
  else:
    return str(test)[0] 
```

```
# Generate Fibonacci sequence to the nth number
def genFib(digits):
  fib = [0, 1]
  if digits == 0:
    return "Enter a whole number larger than 0"
  if digits == 1:
    return [0]
  if digits == 2 :
    return [0 , 1]
  else:
    for i in range(digits-2):
      fib.append(fib[-1] + fib[-2])
    return fib
```

```
import math
def getPrimes(n):
  prime = [True for i in range(n+1)]
  prime[0] = False
  prime[1] = False
  primes = []
  q = 2
  for i in range(math.ceil(math.sqrt(n))):
    if prime[i]:
      j = i**2
      while (j < n+1):
        prime[j] = False
        j = j+i
  for i in range(len(prime)):
    if prime[i] == True:
      primes.append(i)
  return primes
```

```
# Prime Factorization Program
def primeFactors(number):
  factors = []
  primes = getPrimes(number)
  for i in primes:
    while number % i ==0:
      factors.append(i)
      number = number / i
  return factors
```

```
# Classic algos
def collatzConjecture(n):
  steps = 0
  while n != 1:
    if n % 2 == 0:
      n = n/2
      steps += 1
    else:
      n = n*3 +1
      steps +=1
  return steps
```

```
# Longest palindromic substring
# Note this solution times out on leetcode
# We need a faster solution!

def longestPalindrome(s: str) -> str:
    if s == s[::-1]:
        return s
    else:
        longest = s[0]
        for i in range(len(s)):
            for j in range(len(s)):
                s2 = s[i:j+1]
                if s[i:j+1] == s2[::-1] and len(s2) > len(longest):
                    longest = s[i:j+1]
            s3 = s[i+1:j]
            if s[i+1:j] == s3[::-1] and len(s3) > len(longest):
                longest = s[i+1:j]
        return longest
```
